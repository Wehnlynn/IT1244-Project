# Bootstrap
The idea of bootstrapping is to randomly select iid small samples from a big sample. Each small sample can have repeated rows. In the task of finding outliers, we hope to use ML models to learn the features of "Inliers" (datapoints with correct labels), and these models to predict the correct labels of "Outliers". However, it is not possible to use only "Inliers" to train models as we are pretending that we do not have any information of "Outliers".

My idea here is to use bootstrap method to sample 28 small samples. For each of these samples, I train 5 different ML models on them. The full dataset will then be the input for these 5 models. Each model gives 5 different predictions for each datapoint. I then take the average of these predictions as this sample's predictions.

Since I have 28 small samples, I will have 28 samples' predictions. So my final predictions is the average of these 28 sample's predictions. The process can be seen below.

image.png

#### Why this may help us find outliers?
Let's say we have an outlier, $k$ (incorrectly labelled as 1, but should be 0), we now bootstrap 2 samples, called $Sample1$ and $Sample2$. $k$ was chosen to be in $Sample1$ but not $Sample2$.

When we train ML models on $Sample1$, these models will learn the features of $k$. Predicting $k$ again using these models will still result in incorrect class (i.e. the prediction is still 1 instead of 0).

But in $Sample2$, the models do not learn the features of $k$ because $k$ is not in $Sample2$. Thus, using these models to predict $k$ will likely to give us its correct class. (Because these models didn't learn the noise).

By taking the average of these 2 predictions, we take both scenarios into account, and hence we are more likely to obtain the correct labels of the outliers.
